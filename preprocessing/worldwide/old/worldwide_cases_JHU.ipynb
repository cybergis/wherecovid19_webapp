```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
countries = gpd.read_file("World_Countries_Boundaries/World_Countries_Boundaries.shp")
cases = pd.read_csv("05-25-2020.csv")
```

```python
countries
```

```python
cases
```

```python
country_cases = cases.groupby('Country_Region').sum()
country_cases
```

```python
country_cases = country_cases.drop(columns=['FIPS','Lat','Long_'])
country_cases
```

```python
country_cases['Country'] = country_cases.index
country_cases
```

```python
worldwide_cases = pd.merge(countries, country_cases,  how='left', left_on=['CNTRY_NAME'], right_on = ['Country'])
```

```python
worldwide_cases 
```

```python
worldwide_cases.crs
```

```python
worldwide_cases.to_file("worldwide_data.geojson", driver="GeoJSON")
```
