```python
import geopandas as gpd
```

```python
geojson_df = gpd.read_file("World_Countries_Boundaries/World_Countries_Boundaries.shp")
geojson_df
```

```python
geom_simp_series = geojson_df["geometry"].simplify(0.01)
```

```python
geom_simp_df = geom_simp_series.to_frame(name="geometry")
geom_simp_df
```

```python
geojson_df_simp = geojson_df.assign(geometry=geom_simp_df['geometry'])
geojson_df_simp
```

```python
geojson_df_simp.to_file("World_Countries_Boundaries.geojson", driver='GeoJSON')
```
