```python
import pandas as pd
import geopandas as gpd
```

```python
county_attr = pd.read_csv("./us-counties-attributes.csv")
county_attr
```

```python
county_case = gpd.read_file("./nyt_counties_data.geojson")
county_case
```

```python
county_case = county_case[["NAME","state_name","fips","today_case","today_death"]]
county_case['fips'] = county_case['fips'].astype(int) 
county_case
```

```python
county_df = pd.merge(county_case, county_attr, how='right', left_on=['fips'], right_on = ['countyFIPS'])
county_df
```

```python
county_df = county_df[["county","fips","state","state_name","today_case","today_death","TOT_POP","TOT_HH","PCT_CHLDN","PCT_YOUTH","PCT_AD",
                       "PCT_SR","PCT_WHT","PCT_NWHT","MED_HH_INC","PCT_BLW_POV_RT"]]
county_df
```

```python
county_df = county_df[county_df["today_case"].isna() == False]
county_df
```

```python
county_df.to_csv("county_parallel_coordinates.csv")
```
