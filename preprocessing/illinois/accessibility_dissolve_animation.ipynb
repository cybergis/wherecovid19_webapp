```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.colors import ListedColormap
```

```python
chicagoACC0 = gpd.read_file("Chicago_Accessibility_Dissolve_Animation_Test/Chicago_ACC_dissolve0.geojson")
chicagoACC1 = gpd.read_file("Chicago_Accessibility_Dissolve_Animation_Test/Chicago_ACC_dissolve1.geojson")
chicagoACC2 = gpd.read_file("Chicago_Accessibility_Dissolve_Animation_Test/Chicago_ACC_dissolve2.geojson")
chicagoACC3 = gpd.read_file("Chicago_Accessibility_Dissolve_Animation_Test/Chicago_ACC_dissolve3.geojson")
chicagoACC4 = gpd.read_file("Chicago_Accessibility_Dissolve_Animation_Test/Chicago_ACC_dissolve4.geojson")
chicagoACC5 = gpd.read_file("Chicago_Accessibility_Dissolve_Animation_Test/Chicago_ACC_dissolve5.geojson")
chicagoACC6 = gpd.read_file("Chicago_Accessibility_Dissolve_Animation_Test/Chicago_ACC_dissolve6.geojson")
```

```python
chicagoACC0["date"] = "2020-05-01"
chicagoACC1["date"] = "2020-05-02"
chicagoACC2["date"] = "2020-05-03"
chicagoACC3["date"] = "2020-05-04"
chicagoACC4["date"] = "2020-05-05"
chicagoACC5["date"] = "2020-05-06"
chicagoACC6["date"] = "2020-05-07"
```

```python
chicagoACC = pd.concat( [chicagoACC0, chicagoACC1, chicagoACC2, chicagoACC3, chicagoACC4, chicagoACC5, chicagoACC6], axis=0 )
```

```python
chicagoACC
```

```python
chicagoACC = chicagoACC.reset_index()
chicagoACC
```

```python
chicagoACC["dt_start"]="2020-05-01"
chicagoACC["dt_end"]="2020-05-07"
chicagoACC["dt_unit"]="day"
chicagoACC["cases_ts"]="0,0,0,0,0,0,0"
```

```python
chicagoACC
```

```python
chicagoACC = chicagoACC.to_crs({'init': 'epsg:4326'})
```

```python
chicagoACC.to_file("Chicago_ACC_dissolve_animation.geojson", driver="GeoJSON")
```
