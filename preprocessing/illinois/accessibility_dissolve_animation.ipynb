```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.colors import ListedColormap
```

```python
chicagoACC0 = gpd.read_file("Chicago_Acc_2020-05-19_dissolve.geojson")
chicagoACC1 = gpd.read_file("Chicago_ACC_2020-05-20_dissolve.geojson")
chicagoACC2 = gpd.read_file("Chicago_ACC_2020-05-21_dissolve.geojson")

illinoisACC0 = gpd.read_file("Illinois_Acc_2020-05-19_dissolve.geojson")
illinoisACC1 = gpd.read_file("Illinois_ACC_2020-05-20_dissolve.geojson")
illinoisACC2 = gpd.read_file("Illinois_ACC_2020-05-21_dissolve.geojson")
```

```python
chicagoACC0["date"] = "2020-05-19"
chicagoACC1["date"] = "2020-05-20"
chicagoACC2["date"] = "2020-05-21"

illinoisACC0["date"] = "2020-05-19"
illinoisACC1["date"] = "2020-05-20"
illinoisACC2["date"] = "2020-05-21"
```

```python
chicagoACC = pd.concat( [chicagoACC0, chicagoACC1, chicagoACC2], axis=0 )
illinoisACC = pd.concat( [illinoisACC0, illinoisACC1, illinoisACC2], axis=0 )
```

```python
chicagoACC = chicagoACC.reset_index()
illinoisACC = illinoisACC.reset_index()
```

```python
chicagoACC["dt_start"]="2020-05-19"
chicagoACC["dt_end"]="2020-05-21"
chicagoACC["dt_unit"]="day"
chicagoACC["cases_ts"]="0,0,0"

illinoisACC["dt_start"]="2020-05-19"
illinoisACC["dt_end"]="2020-05-21"
illinoisACC["dt_unit"]="day"
illinoisACC["cases_ts"]="0,0,0"
```

```python
chicagoACC = chicagoACC.to_crs({'init': 'epsg:4326'})
illinoisACC = illinoisACC.to_crs({'init': 'epsg:4326'})
```

```python
chicagoACC.to_file("Chicago_ACC_dissolve_animation.geojson", driver="GeoJSON")
illinoisACC.to_file("Illinois_ACC_dissolve_animation.geojson", driver="GeoJSON")
```
